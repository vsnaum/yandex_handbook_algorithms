### <b>A + B</b>
Ğ—Ğ°Ğ´Ğ°Ñ‡Ğ°-Ñ€Ğ°Ğ·Ğ¼Ğ¸Ğ½ĞºĞ°: Ğ½Ğ°Ğ¹Ğ´Ğ¸Ñ‚Ğµ ÑÑƒĞ¼Ğ¼Ñƒ Ğ´Ğ²ÑƒÑ… Ñ‡Ğ¸ÑĞµĞ».

<b>Ğ¤Ğ¾Ñ€Ğ¼Ğ°Ñ‚ Ğ²Ğ²Ğ¾Ğ´Ğ°</b>

Ğ¦ĞµĞ»Ñ‹Ğµ Ñ‡Ğ¸ÑĞ»Ğ° ğ‘ Ğ¸ ğ‘ Ğ½Ğ° Ğ¾Ğ´Ğ½Ğ¾Ğ¹ ÑÑ‚Ñ€Ğ¾ĞºĞµ (Ñ€Ğ°Ğ·Ğ´ĞµĞ»Ñ‘Ğ½Ğ½Ñ‹Ğµ Ğ¿Ñ€Ğ¾Ğ±ĞµĞ»Ğ¾Ğ¼).

ĞĞ³Ñ€Ğ°Ğ½Ğ¸Ñ‡ĞµĞ½Ğ¸Ñ: 0 â‰¤ ğ‘, ğ‘ â‰¤ 9.

Ğ¡ÑƒĞ¼Ğ¼Ğ° ğ‘ Ğ¸ ğ‘.

```python
def inp_list(fname):
    with open(fname,'r') as f:
        l = f.readlines()[0].split(' ')
        for i in range(len(l)):
            l[i] = int(l[i])
    return l

l = inp_list('input.txt')
print(l[0]+l[1])
```

### <b>A(x) + B(x)</b>
Ğ ĞµÑˆĞ¸Ñ‚Ğµ Ğ½ĞµĞ¼Ğ½Ğ¾Ğ³Ğ¾ Ğ±Ğ¾Ğ»ĞµĞµ ÑĞ»Ğ¾Ğ¶Ğ½ÑƒÑ Ğ·Ğ°Ğ´Ğ°Ñ‡Ñƒ.

ĞĞµĞ¾Ğ±Ñ…Ğ¾Ğ´Ğ¸Ğ¼Ğ¾ Ğ²Ñ‹Ñ‡Ğ¸ÑĞ»Ğ¸Ñ‚ÑŒ ÑÑƒĞ¼Ğ¼Ñƒ Ğ´Ğ²ÑƒÑ… Ğ¼Ğ½Ğ¾Ğ³Ğ¾Ñ‡Ğ»ĞµĞ½Ğ¾Ğ²

<b>Ğ¤Ğ¾Ñ€Ğ¼Ğ°Ñ‚ Ğ²Ğ²Ğ¾Ğ´Ğ°</b>

Ğ’ Ğ¿ĞµÑ€Ğ²Ğ¾Ğ¹ ÑÑ‚Ñ€Ğ¾ĞºĞµ Ğ·Ğ°Ğ¿Ğ¸ÑĞ°Ğ½Ğ¾ Ğ¾Ğ´Ğ½Ğ¾ Ñ†ĞµĞ»Ğ¾Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾ 
ğ‘› (0 â‰¤ ğ‘› â‰¤ 10). Ğ’ Ğ²Ñ‚Ğ¾Ñ€Ğ¾Ğ¹ ÑÑ‚Ñ€Ğ¾ĞºĞµ Ğ·Ğ°Ğ¿Ğ¸ÑĞ°Ğ½Ñ‹ Ñ‡Ğ¸ÑĞ»Ğ° ğ‘_ğ‘›, ğ‘_ğ‘›âˆ’1, â€¦, ğ‘_0 (âˆ’100 â‰¤ ğ‘_ğ‘– â‰¤ 100, ğ‘_ğ‘› â‰  0).


Ğ’ Ñ‚Ñ€ĞµÑ‚ÑŒĞµĞ¹ ÑÑ‚Ñ€Ğ¾ĞºĞµ Ğ·Ğ°Ğ¿Ğ¸ÑĞ°Ğ½Ğ¾ Ğ¾Ğ´Ğ½Ğ¾ Ñ†ĞµĞ»Ğ¾Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾ ğ‘š (0 â‰¤ ğ‘š â‰¤ 10). Ğ’ Ñ‡ĞµÑ‚Ğ²ĞµÑ€Ñ‚Ğ¾Ğ¹ ÑÑ‚Ñ€Ğ¾ĞºĞµ Ğ·Ğ°Ğ¿Ğ¸ÑĞ°Ğ½Ñ‹ Ñ‡Ğ¸ÑĞ»Ğ° ğ‘_ğ‘š, ğ‘_ğ‘šâˆ’1, â€¦, ğ‘_0 (âˆ’100 â‰¤ ğ‘_ğ‘– â‰¤ 100, ğ‘_ğ‘š â‰  0).

Ğ“Ğ°Ñ€Ğ°Ğ½Ñ‚Ğ¸Ñ€ÑƒĞµÑ‚ÑÑ, Ñ‡Ñ‚Ğ¾ ğ‘_ğ‘› + ğ‘_ğ‘š â‰  0.

<b>Ğ¤Ğ¾Ñ€Ğ¼Ğ°Ñ‚ Ğ²Ñ‹Ğ²Ğ¾Ğ´Ğ°</b>

Ğ’ Ğ¿ĞµÑ€Ğ²Ğ¾Ğ¹ ÑÑ‚Ñ€Ğ¾ĞºĞµ Ğ²Ñ‹Ğ²ĞµĞ´Ğ¸Ñ‚Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾ ğ‘˜ â€” ÑÑ‚ĞµĞ¿ĞµĞ½ÑŒ Ğ¼Ğ½Ğ¾Ğ³Ğ¾Ñ‡Ğ»ĞµĞ½Ğ° ğ´(ğ‘¥) + ğµ(ğ‘¥).

Ğ’ Ğ²Ñ‚Ğ¾Ñ€Ğ¾Ğ¹ ÑÑ‚Ñ€Ğ¾ĞºĞµ Ğ²Ñ‹Ğ²ĞµĞ´Ğ¸Ñ‚Ğµ ĞºĞ¾ÑÑ„Ñ„Ğ¸Ñ†Ğ¸ĞµĞ½Ñ‚Ñ‹ ÑÑ‚Ğ¾Ğ³Ğ¾ Ğ¼Ğ½Ğ¾Ğ³Ğ¾Ñ‡Ğ»ĞµĞ½Ğ°: 
ğ‘_ğ‘˜, ğ‘_ğ‘˜âˆ’1, â€¦, ğ‘_1, ğ‘_0.

```python
def inp_list(fname,line):
    with open(fname,'r') as f:
        l = f.readlines()[line].split(' ')
        for i in range(len(l)):
            l[i] = int(l[i])
    return l

def print_list(l):
    for e in l:
        print(e,end=' ')

A = inp_list('input.txt',1)[::-1]
B = inp_list('input.txt',3)[::-1]

S = []

for i in range(min(len(A),len(B))):
    S.append(A[i] + B[i])
S+=A[i+1:]
S+=B[i+1:]


print(len(S)-1)
print_list(S[::-1])
```

### <b>A+B ÑÑ‚Ñ€Ğ¾ĞºĞ¸</b>

ĞĞµĞ¾Ğ±Ñ…Ğ¾Ğ´Ğ¸Ğ¼Ğ¾ Ğ²Ñ‹Ñ‡Ğ¸ÑĞ»Ğ¸Ñ‚ÑŒ "ĞºÑ€Ğ¸Ğ²ÑƒÑ" ÑÑƒĞ¼Ğ¼Ñƒ Ğ´Ğ²ÑƒÑ… ÑÑ‚Ñ€Ğ¾Ğº ğ´ Ğ¸ ğµ Ğ¾Ğ´Ğ¸Ğ½Ğ°ĞºĞ¾Ğ²Ğ¾Ğ¹ Ğ´Ğ»Ğ¸Ğ½Ñ‹.

"ĞšÑ€Ğ¸Ğ²Ğ¾Ğ¹" ÑÑƒĞ¼Ğ¼Ğ¾Ğ¹ Ğ´Ğ²ÑƒÑ… ÑÑ‚Ñ€Ğ¾Ğº Ğ½Ğ°Ğ·Ñ‹Ğ²Ğ°ĞµÑ‚ÑÑ Ğ¾Ğ¿ĞµÑ€Ğ°Ñ†Ğ¸Ñ ÑĞ»ĞµĞ´ÑƒÑÑ‰ĞµĞ³Ğ¾ Ğ²Ğ¸Ğ´Ğ°:

ğ¶ = ğ´ + ğµ = ğ´_1ğµ_1ğ´_2ğµ_2â€¦ğ´_ğ‘›âˆ’1ğµ_ğ‘›âˆ’1ğ´_ğ‘›ğµ_ğ‘›

<b>Ğ¤Ğ¾Ñ€Ğ¼Ğ°Ñ‚ Ğ²Ğ²Ğ¾Ğ´Ğ°</b>

Ğ’ Ğ¿ĞµÑ€Ğ²Ğ¾Ğ¹ ÑÑ‚Ñ€Ğ¾ĞºĞµ Ğ·Ğ°Ğ¿Ğ¸ÑĞ°Ğ½Ğ¾ Ñ†ĞµĞ»Ğ¾Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾ ğ‘› (1â‰¤ğ‘›â‰¤10).

Ğ’Ñ‚Ğ¾Ñ€Ğ°Ñ ÑÑ‚Ñ€Ğ¾ĞºĞ° ÑĞ¾Ğ´ĞµÑ€Ğ¶Ğ¸Ñ‚ ÑÑ‚Ñ€Ğ¾ĞºÑƒ ğ´. (ğ´_ğ‘– âˆˆ {ğ‘,ğ‘,â€¦,ğ‘§}).

Ğ¢Ñ€ĞµÑ‚ÑŒÑ ÑÑ‚Ñ€Ğ¾ĞºĞ° ÑĞ¾Ğ´ĞµÑ€Ğ¶Ğ¸Ñ‚ ÑÑ‚Ñ€Ğ¾ĞºÑƒ ğµ. (ğµ_ğ‘– âˆˆ {ğ‘,ğ‘,â€¦,ğ‘§}).

Ğ’Ñ‹Ğ²ĞµĞ´Ğ¸Ñ‚Ğµ Ğ¾Ğ´Ğ½Ñƒ ÑÑ‚Ñ€Ğ¾ĞºÑƒ, ÑĞ¾Ğ´ĞµÑ€Ğ¶Ğ°Ñ‰ÑƒÑ ÑÑ‚Ñ€Ğ¾ĞºÑƒ Ğ¡ = ğ´ + ğµ.

```python
def inp_str(fname,line):
    with open(fname,'r') as f:
        l = f.readlines()[line]
    return l

A = inp_str('input.txt',1)
B = inp_str('input.txt',2)

s=''
for i in range(len(A)):
    s += A[i]+B[i]

print(s)
```

### <b>A+B Ğ¼Ğ°Ñ‚Ñ€Ğ¸Ñ†Ñ‹</b>
ĞĞµĞ¾Ğ±Ñ…Ğ¾Ğ´Ğ¸Ğ¼Ğ¾ Ğ²Ñ‹Ñ‡Ğ¸ÑĞ»Ğ¸Ñ‚ÑŒ ÑÑƒĞ¼Ğ¼Ñƒ Ğ´Ğ²ÑƒÑ… Ğ¼Ğ°Ñ‚Ñ€Ğ¸Ñ† ğ¶ = ğ´ + ğµ.

<b>Ğ¤Ğ¾Ñ€Ğ¼Ğ°Ñ‚ Ğ²Ğ²Ğ¾Ğ´Ğ°</b>

Ğ’ Ğ¿ĞµÑ€Ğ²Ğ¾Ğ¹ ÑÑ‚Ñ€Ğ¾ĞºĞµ Ğ·Ğ°Ğ¿Ğ¸ÑĞ°Ğ½Ğ¾ Ğ´Ğ²Ğ° Ñ†ĞµĞ»Ñ‹Ñ… Ñ‡Ğ¸ÑĞ»Ğ° ğ‘› Ğ¸ ğ‘š (1â‰¤ğ‘›,ğ‘šâ‰¤10).

Ğ”Ğ°Ğ»ĞµĞµ Ğ¸Ğ´ÑƒÑ‚ ğ‘› ÑÑ‚Ñ€Ğ¾Ğº, ĞºĞ°Ğ¶Ğ´Ğ°Ñ ÑĞ¾Ğ´ĞµÑ€Ğ¶Ğ¸Ñ‚ Ğ¿Ğ¾ ğ‘š Ñ‡Ğ¸ÑĞµĞ», -- Ğ¾Ğ¿Ğ¸ÑĞ°Ğ½Ğ¸Ğµ ÑĞ»ĞµĞ¼ĞµĞ½Ñ‚Ğ¾Ğ² Ğ¼Ğ°Ñ‚Ñ€Ğ¸Ñ†Ñ‹ ğ´ (-100 â‰¤ ğ´_ğ‘–,ğ‘— â‰¤ 100).

Ğ”Ğ°Ğ»ĞµĞµ Ğ¸Ğ´ÑƒÑ‚ ğ‘› ÑÑ‚Ñ€Ğ¾Ğº, ĞºĞ°Ğ¶Ğ´Ğ°Ñ ÑĞ¾Ğ´ĞµÑ€Ğ¶Ğ¸Ñ‚ Ğ¿Ğ¾ ğ‘š Ñ‡Ğ¸ÑĞµĞ», -- Ğ¾Ğ¿Ğ¸ÑĞ°Ğ½Ğ¸Ğµ ÑĞ»ĞµĞ¼ĞµĞ½Ñ‚Ğ¾Ğ² Ğ¼Ğ°Ñ‚Ñ€Ğ¸Ñ†Ñ‹ ğµ (âˆ’100 â‰¤ ğµ_ğ‘–,ğ‘— â‰¤ 100).

<b>Ğ¤Ğ¾Ñ€Ğ¼Ğ°Ñ‚ Ğ²Ñ‹Ğ²Ğ¾Ğ´Ğ°</b>

Ğ’Ñ‹Ğ²ĞµĞ´Ğ¸Ñ‚Ğµ ğ‘› ÑÑ‚Ñ€Ğ¾Ğº, ĞºĞ°Ğ¶Ğ´Ğ°Ñ ÑĞ¾Ğ´ĞµÑ€Ğ¶Ğ°Ñ‰Ğ°Ñ Ğ¿Ğ¾ ğ‘š Ñ‡Ğ¸ÑĞµĞ», -- Ğ¾Ğ¿Ğ¸ÑĞ°Ğ½Ğ¸Ğµ ÑĞ»ĞµĞ¼ĞµĞ½Ñ‚Ğ¾Ğ² Ğ¼Ğ°Ñ‚Ñ€Ğ¸Ñ†Ñ‹ Ğ¡.

```python
def inp_mat(n_mat):
    with open('input.txt','r') as f:
        ll = f.readlines()
        size = tuple(int(i) for i in ll[0].split(' '))
        mats = []
        line = 1
        for _ in range(n_mat):
            m = []
            for _ in range(size[0]):
                tmp = list(int(i) for i in ll[line].split(' '))
                m.append( tmp )
                line+=1
            mats.append(m)
        return mats
        
def print_mat(M):
    for r in range(len(M)):
        for c in range(len(M[0])):
            print(M[r][c],end = ' ')
        print('')

mats = inp_mat(2)
A = mats[0]
B = mats[1]

res = []

for r in range(len(A)):
    res.append([])
    for c in range(len(A[0])):
        res[r].append(A[r][c]+B[r][c])
print_mat(res)
```
