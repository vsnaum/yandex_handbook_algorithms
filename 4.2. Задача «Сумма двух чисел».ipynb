### <b>A + B</b>
Задача-разминка: найдите сумму двух чисел.

<b>Формат ввода</b>

Целые числа 𝑎 и 𝑏 на одной строке (разделённые пробелом).

Ограничения: 0 ≤ 𝑎, 𝑏 ≤ 9.

<b>Формат вывода</b>

Сумма 𝑎 и 𝑏.

```python
def inp_list(fname):
    with open(fname,'r') as f:
        l = f.readlines()[0].split(' ')
        for i in range(len(l)):
            l[i] = int(l[i])
    return l

l = inp_list('input.txt')
print(l[0]+l[1])
```

### <b>A(x) + B(x)</b>
Решите немного более сложную задачу.

Необходимо вычислить сумму двух многочленов 
𝐴(𝑥) = 𝑎_𝑛 ⋅ 𝑥^𝑛 + … + 𝑎_1 ⋅ 𝑥 + 𝑎_0  и 𝐵(𝑥) = 𝑏_𝑚 ⋅ 𝑥_𝑚 + … + 𝑏_1 ⋅ 𝑥 + 𝑏_0.

<b>Формат ввода</b>

В первой строке записано одно целое число 
𝑛 (0 ≤ 𝑛 ≤ 10). В второй строке записаны числа 𝑎_𝑛, 𝑎_𝑛−1, …, 𝑎_0 (−100 ≤ 𝑎_𝑖 ≤ 100, 𝑎_𝑛 ≠ 0).


В третьей строке записано одно целое число 𝑚 (0 ≤ 𝑚 ≤ 10). В четвертой строке записаны числа 𝑏_𝑚, 𝑏_𝑚−1, …, 𝑏_0 (−100 ≤ 𝑏_𝑖 ≤ 100, 𝑏_𝑚 ≠ 0).

Гарантируется, что 𝑎_𝑛 + 𝑏_𝑚 ≠ 0.

<b>Формат вывода</b>

В первой строке выведите число 𝑘 — степень многочлена 𝐴(𝑥) + 𝐵(𝑥).

В второй строке выведите коэффициенты этого многочлена: 
𝑐_𝑘, 𝑐_𝑘−1, …, 𝑐_1, 𝑐_0.

```python
def inp_list(fname,line):
    with open(fname,'r') as f:
        l = f.readlines()[line].split(' ')
        for i in range(len(l)):
            l[i] = int(l[i])
    return l

def print_list(l):
    for e in l:
        print(e,end=' ')

A = inp_list('input.txt',1)[::-1]
B = inp_list('input.txt',3)[::-1]

S = []

for i in range(min(len(A),len(B))):
    S.append(A[i] + B[i])
S+=A[i+1:]
S+=B[i+1:]


print(len(S)-1)
print_list(S[::-1])
```

### <b>A+B строки</b>

Необходимо вычислить "кривую" сумму двух строк 𝐴 и 𝐵 одинаковой длины.

"Кривой" суммой двух строк называется операция следующего вида:

𝐶 = 𝐴 + 𝐵 = 𝐴_1𝐵_1𝐴_2𝐵_2…𝐴_𝑛−1𝐵_𝑛−1𝐴_𝑛𝐵_𝑛
 
Где 𝑛 -- длина строк 𝐴 и 𝐵.

<b>Формат ввода</b>

В первой строке записано целое число 𝑛 (1≤𝑛≤10).

Вторая строка содержит строку 𝐴. (𝐴_𝑖 ∈ {𝑎,𝑏,…,𝑧}).

Третья строка содержит строку 𝐵. (𝐵_𝑖 ∈ {𝑎,𝑏,…,𝑧}).

<b>Формат вывода</b>

Выведите одну строку, содержащую строку С = 𝐴 + 𝐵.

```python
def inp_str(fname,line):
    with open(fname,'r') as f:
        l = f.readlines()[line]
    return l

A = inp_str('input.txt',1)
B = inp_str('input.txt',2)

s=''
for i in range(len(A)):
    s += A[i]+B[i]

print(s)
```

### <b>A+B матрицы</b>
Необходимо вычислить сумму двух матриц 𝐶 = 𝐴 + 𝐵.

<b>Формат ввода</b>

В первой строке записано два целых числа 𝑛 и 𝑚 (1≤𝑛,𝑚≤10).

Далее идут 𝑛 строк, каждая содержит по 𝑚 чисел, -- описание элементов матрицы 𝐴 (-100 ≤ 𝐴_𝑖,𝑗 ≤ 100).

Далее идут 𝑛 строк, каждая содержит по 𝑚 чисел, -- описание элементов матрицы 𝐵 (−100 ≤ 𝐵_𝑖,𝑗 ≤ 100).

<b>Формат вывода</b>

Выведите 𝑛 строк, каждая содержащая по 𝑚 чисел, -- описание элементов матрицы С.

```python
def inp_mat(n_mat):
    with open('input.txt','r') as f:
        ll = f.readlines()
        size = tuple(int(i) for i in ll[0].split(' '))
        mats = []
        line = 1
        for _ in range(n_mat):
            m = []
            for _ in range(size[0]):
                tmp = list(int(i) for i in ll[line].split(' '))
                m.append( tmp )
                line+=1
            mats.append(m)
        return mats
        
def print_mat(M):
    for r in range(len(M)):
        for c in range(len(M[0])):
            print(M[r][c],end = ' ')
        print('')

mats = inp_mat(2)
A = mats[0]
B = mats[1]

res = []

for r in range(len(A)):
    res.append([])
    for c in range(len(A[0])):
        res[r].append(A[r][c]+B[r][c])
print_mat(res)
```
