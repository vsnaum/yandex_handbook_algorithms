```python
def inp_list(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line].split(' ')
        for i in range(len(l)):
            l[i] = int(l[i])
    return l


def MaxPairwiseProduct(A):
    m1 = A[0]
    m2 = A[1]
    if m2 > m1:
        m1, m2 = m2, m1

    for i in range(2,len(A)):
        if A[i] > m1:
            m2, m1 = m1, A[i]
        else:
            if A[i] > m2:
                m2 = A[i]
    return m1 * m2

nums = inp_list(1)

print(MaxPairwiseProduct(nums))
```

```python
import random as rnd

def inp_str(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line]
    return l

def max_prod_count_if(A):
    m1 = A[0]
    m2 = A[1]
    count_if = 1
    if m2 > m1:
        m1, m2 = m2, m1
    for i in range(2,len(A)):
        count_if +=1
        if A[i] > m1:
            m2, m1 = m1, A[i]
        else:
            count_if +=1
            if A[i] > m2:
                m2 = A[i]
    return count_if


def gen_int_list(n,min_val,max_val):
    l = []
    for _ in range(n):
        l.append(rnd.randint(min_val,max_val))
    return l

def print_list(l):
    for e in l:
        print(e,end=' ')

n = int(inp_str(0))
run = True
i=0
while run and i<1000:
    i+=1
    l = gen_int_list(n,0,200000)
    count_if = max_prod_count_if(l)
    if count_if > 1.5*n:
        run = False


if run:
    print('No')
else:
    print('Yes')
    print_list(l)
```
