### <b>Максимальное произведение</b>
Дана последовательность неотрицательных целых чисел 𝑎_1,…,𝑎_𝑛. Вычислите max 𝑎_𝑖 ⋅ 𝑎_𝑗, 1 ≤ 𝑖≠𝑗 ≤ 𝑛.

Обратите внимание, что 𝑖 и 𝑗 должны быть разными, хотя в каких-то случаях можно наблюдать, что 𝑎_𝑖 = 𝑎_𝑗.

<b>Формат ввода</b>

Первая строка содержит целое число 𝑛. Следующая строка содержит 𝑛 неотрицательных целых чисел 𝑎_1,…,𝑎_𝑛 (разделены пробелами).

Ограничения: 2≤𝑛≤2⋅10^5 ; 0≤𝑎_1,…,𝑎_𝑛≤2⋅10^5.

<b>Формат вывода</b>

Максимальное попарное произведение.

```python
def inp_list(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line].split(' ')
        for i in range(len(l)):
            l[i] = int(l[i])
    return l


def MaxPairwiseProduct(A):
    m1 = A[0]
    m2 = A[1]
    if m2 > m1:
        m1, m2 = m2, m1

    for i in range(2,len(A)):
        if A[i] > m1:
            m2, m1 = m1, A[i]
        else:
            if A[i] > m2:
                m2 = A[i]
    return m1 * m2

nums = inp_list(1)

print(MaxPairwiseProduct(nums))
```

<b>Формат ввода </b>

Целое число 𝑛
n.

Ограничения: 2≤𝑛≤200000.

<b>Формат вывода</b>

В единственной строке выведите No, если подходящих входных данных не существует. Иначе в первой строке выведите Yes, а во второй строке 𝑛 чисел 𝑎_1,𝑎_2,…,𝑎_𝑛 (0≤𝑎_𝑖≤200000) — найденный контрпример.

Если походящих последовательностей несколько, выведите любую из них.

```python
import random as rnd

def inp_str(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line]
    return l

def max_prod_count_if(A):
    m1 = A[0]
    m2 = A[1]
    count_if = 1
    if m2 > m1:
        m1, m2 = m2, m1
    for i in range(2,len(A)):
        count_if +=1
        if A[i] > m1:
            m2, m1 = m1, A[i]
        else:
            count_if +=1
            if A[i] > m2:
                m2 = A[i]
    return count_if


def gen_int_list(n,min_val,max_val):
    l = []
    for _ in range(n):
        l.append(rnd.randint(min_val,max_val))
    return l

def print_list(l):
    for e in l:
        print(e,end=' ')

n = int(inp_str(0))
run = True
i=0
while run and i<1000:
    i+=1
    l = gen_int_list(n,0,200000)
    count_if = max_prod_count_if(l)
    if count_if > 1.5*n:
        run = False


if run:
    print('No')
else:
    print('Yes')
    print_list(l)
```

### <b>Максимальное произведение трех чисел</b>
Дана последовательность целых чисел 𝑎_1,…,𝑎_𝑛. Вычислите

max  𝑎_𝑖 ⋅ 𝑎_𝑗 ⋅ 𝑎_𝑘 (1≤𝑖≠𝑗,𝑖≠𝑘,𝑗≠𝑘≤𝑛).

Обратите внимание, что индексы 𝑖,𝑗j и 𝑘 должны быть различны. Хотя в каких-то случаях значения элементов могут совпадать, то есть, например, 𝑎_𝑖=𝑎_𝑗.

<b>Формат ввода</b>

Первая строка содержит целое число 𝑛. Следующая строка содержит 𝑛 целых чисел 𝑎_1,…,𝑎_𝑛 (разделены пробелами).

Ограничения: 
3≤𝑛≤2⋅10^5 ; −2⋅10^5≤𝑎_1,…,𝑎_𝑛≤2⋅10^5.

<b>Формат вывода</b>

Максимальное произведение трех элементов.

```python
def inp_list(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line].split(' ')
        for i in range(len(l)):
            l[i] = int(l[i])
    return l

def max_prod3_naive(l):
    prod = -10**16
    for i in range(len(l)):
        for j in range(i+1,len(l)):
            for k in range(j+1,len(l)):
                if i != j != k:
                    prod_new = l[i]*l[j]*l[k]
                    if prod_new > prod:
                        prod = prod_new
    return prod

def max_prod3(l):
    l.sort(reverse=True)
    m1 = l[0]*l[1]*l[2]
    m2 = l[-1]*l[-2]*l[0]
    return max(m1,m2)
    
l = inp_list(1)

print(max_prod3(l))
```

### <b>Максимальное произведение четырех чисел</b>

Дана последовательность целых чисел 𝑎_1,…,𝑎_𝑛. Вычислите

max 𝑎_𝑖 ⋅ 𝑎_𝑗 ⋅ 𝑎_𝑘 ⋅ 𝑎_𝑙 (1≤𝑖≠𝑗≠𝑘≠𝑙≤𝑛).

Обратите внимание, что индексы  𝑖,𝑗,𝑘 и 𝑙 должны быть различны. Хотя в каких-то случаях значения элементов могут совпадать, то есть, например, 𝑎_𝑖=𝑎_𝑙.

<b>Формат ввода</b>

Первая строка содержит целое число 𝑛. Следующая строка содержит 𝑛 целых чисел 𝑎_1,…,𝑎_𝑛  (разделены пробелами).

Ограничения: 
4≤𝑛≤2⋅10^5 ; −2⋅10^4≤𝑎_1,…,𝑎_𝑛≤2⋅10^4.

<b>Формат вывода</b>

Максимальное произведение четырех элементов.

```python
def inp_list(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line].split(' ')
        for i in range(len(l)):
            l[i] = int(l[i])
    return l


def max_prod4(l):
    l.sort()
    
    m1 = l[0]*l[1]*l[2]*l[3]
    m2 = l[-1]*l[-2]*l[-3]*l[-4]
    m3 = l[0]*l[1]*l[-1]*l[-2]
    
    return max(m1,m2,m3)


ll = inp_list(1)

print(max_prod4(ll))
```
