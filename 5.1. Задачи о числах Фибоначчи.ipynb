```python
def inp_str(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line]
    return l

def inp_list(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line].split(' ')
        for i in range(len(l)):
            l[i] = int(l[i])
    return l

def PisanoPeriod(m):
    current = 0
    next = 1
    period = 0
    while True:
        current, next = next, ((current + next) % m)
        period = period + 1
        if current == 0 and next == 1:
            return period

def fib2(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        fib_n2 = 0
        fib_n1 = 1
        fib_res = 1
        for _ in range(2,n):
            fib_n2 = fib_n1
            fib_n1 = fib_res
            fib_res = fib_n1+fib_n2        
    return fib_res

def fib_mod_m(n_fib,m):
    new_n = n_fib % PisanoPeriod(m)
    return (fib2(new_n) % m)

n,m = inp_list(0)
print(fib_mod_m(n,m))
```

### <b>–ü–æ—Å–ª–µ–¥–Ω—è—è —Ü–∏—Ñ—Ä–∞ —Å—É–º–º—ã —á–∏—Å–µ–ª –§–∏–±–æ–Ω–∞—á—á–∏ </b>
–î–∞–Ω–æ —á–∏—Å–ª–æ ùëõ, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–π—Ç–∏ –ø–æ—Å–ª–µ–¥–Ω—é—é —Ü–∏—Ñ—Ä—É —Å—É–º–º—ã  ùêπ_0 + ùêπ_1 + ùêπ_2 + ‚Ä¶ + ùêπ_ùëõ


<b>–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞</b>

–¶–µ–ª–æ–µ —á–∏—Å–ª–æ ùëõ.

<b>–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è: </b> 

0‚â§ùëõ‚â§10^14

<b>–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞</b>

–ü–æ—Å–ª–µ–¥–Ω—è—è —Ü–∏—Ñ—Ä–∞ —Å—É–º–º—ã ùêπ_0 + ùêπ_1 + ùêπ_2 + ‚Ä¶ + ùêπ_ùëõ.

```python
# –ü–æ—Å–ª–µ–¥–Ω—è—è —Ü–∏—Ñ—Ä–∞ —Å—É–º–º—ã —á–∏—Å–µ–ª –§–∏–±–æ–Ω–∞—á—á–∏

def inp_str(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line]
    return l

def fib2(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        fib_n2 = 0
        fib_n1 = 1
        fib_res = 1
        for _ in range(2,n):
            fib_n2 = fib_n1
            fib_n1 = fib_res
            fib_res = fib_n1+fib_n2        
    return fib_res

def s_last_naive(n):
    return (fib2(n+2) - 1) %10

# —Å–ª–∏—à–∫–æ–º –º–µ–¥–ª–µ–Ω–Ω–æ –¥–ª—è —Ç–µ—Å—Ç–∞ 48
def s_last_slow(n):
    if n <= 2:
        return n
    else:
        fib_n2 = 0
        fib_n1 = 1
        fib_res = 1
        for _ in range(2,n+2):
            fib_n2 = fib_n1
            fib_n1 = fib_res
            fib_res = (fib_n1+fib_n2)%10       
    if fib_res == 0:
        return 9
    return fib_res - 1

def PisanoPeriod(m):
    current = 0
    next = 1
    period = 0
    while True:
        current, next = next, ((current + next) % m)
        period = period + 1
        if current == 0 and next == 1:
            return period

def fib_mod_m(n_fib,m):
    new_n = n_fib % PisanoPeriod(m)
    return (fib2(new_n) % m)

def s_last(n):
    if n <= 2:
        return n
    res = fib_mod_m(n+2,10)    
    if res == 0:
        return 9
    return res - 1

def stress_test():
    for n in range(0,51):
        print(f"n={n}")
        naive = s_last_naive(n)
        elegant = s_last(n)
        assert naive == elegant, f"naive: {naive}  elegant: {elegant}"


n = int(inp_str(0))
print(s_last(n))
```

### –ü–æ—Å–ª–µ–¥–Ω—è—è —Ü–∏—Ñ—Ä–∞ —á–∞—Å—Ç–∏—á–Ω–æ–π —Å—É–º–º—ã —á–∏—Å–µ–ª –§–∏–±–æ–Ω–∞—á—á–∏
–î–∞–Ω—ã —á–∏—Å–ª–∞ ùëö –∏ ùëõ, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–π—Ç–∏ –ø–æ—Å–ª–µ–¥–Ω—é—é —Ü–∏—Ñ—Ä—É —Å—É–º–º—ã ùêπ_ùëö + ùêπ_ùëö+1 + ‚Ä¶ + ùêπ_ùëõ

<b>–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞</b>

<b>–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è: </b>

0‚â§ùëö‚â§ùëõ‚â§10^14

<b>–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞</b>

–ü–æ—Å–ª–µ–¥–Ω—è—è —Ü–∏—Ñ—Ä–∞ —Å—É–º–º—ã ùêπ_ùëö + ùêπ_ùëö+1 + ‚Ä¶ + ùêπ_ùëõ

```python
# –ü–æ—Å–ª–µ–¥–Ω—è—è —Ü–∏—Ñ—Ä–∞ —á–∞—Å—Ç–∏—á–Ω–æ–π —Å—É–º–º—ã —á–∏—Å–µ–ª –§–∏–±–æ–Ω–∞—á—á–∏

def inp_list(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line].split(' ')
        for i in range(len(l)):
            l[i] = int(l[i])
    return l

def fib2(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        fib_n2 = 0
        fib_n1 = 1
        fib_res = 1
        for _ in range(2,n):
            fib_n2 = fib_n1
            fib_n1 = fib_res
            fib_res = fib_n1+fib_n2        
    return fib_res


def PisanoPeriod(m):
    current = 0
    next = 1
    period = 0
    while True:
        current, next = next, ((current + next) % m)
        period = period + 1
        if current == 0 and next == 1:
            return period


def ps_last(m,n):
    pp = PisanoPeriod(10)
    new_m = m % pp
    new_n = n % pp
    if new_n < new_m:
        new_n += pp

    s_m = fib2(new_m-1+2) - 1
    s_n = fib2(new_n+2) - 1
    ss = s_n - s_m
    
    return ss % 10

m,n = inp_list(0)


print(ps_last(m,n))
```
