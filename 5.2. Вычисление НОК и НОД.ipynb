```python
def inp_list(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line].split(' ')
        for i in range(len(l)):
            l[i] = int(l[i])
    return l


def GCD(a,b):
    while a>0 and b>0:
        if a>=b:
            a = a % b
        else:
            b = b % a
    return max(a,b)


a,b = inp_list(0)

print(GCD(a,b))
```

### <b>–ù–∞–∏–±–æ–ª—å—à–µ–µ —á–∏—Å–ª–æ —à–∞–≥–æ–≤ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –ï–≤–∫–ª–∏–¥–∞</b>

–ê–ª–≥–æ—Ä–∏—Ç–º –ï–≤–∫–ª–∏–¥–∞ –º–æ–∂–Ω–æ –∑–∞–ø–∏—Å–∞—Ç—å –≤ —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ–π —Ñ–æ—Ä–º–µ:

GCD(a, b):
    if a = 0 or b = 0:

    return GCD(b,a mod b)

–í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ GCD(5,0) –Ω–µ –ø–æ—Ä–æ–∂–¥–∞–µ—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö —Ä–µ–∫—É—Ä—Å–∏–≤–Ω—ã—Ö –≤—ã–∑–æ–≤–æ–≤. –ê –≤—ã–∑–æ–≤ GCD(5,2) - –ø–æ—Ä–æ–∂–¥–∞–µ—Ç –≤—ã–∑–æ–≤—ã GCD(2,1) –∏ GCD(1,0).

–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –ï–≤–∫–ª–∏–¥–∞ –±—É–¥–µ—Ç —Ä–∞–≤–Ω—è—Ç—å—Å—è –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –≤—ã–∑–æ–≤–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ GCD –¥–ª—è –∑–∞–¥–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π ùëé –∏ ùëè.

–î–∞–Ω–æ —á–∏—Å–ª–æ ùëõ, –Ω–∞–π–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è ùëé –∏ ùëè (0‚â§ùëé,ùëè‚â§ùëõ), –¥–ª—è –∫–æ—Ç–æ—Ä—ã—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–∑–æ–≤–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ GCD –±—É–¥–µ—Ç –Ω–∞–∏–±–æ–ª—å—à–∏–º.

<b>–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞</b>

–¶–µ–ª–æ–µ —á–∏—Å–ª–æ ùëõ.

<b>–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è:</b> 

1‚â§ùëõ‚â§10^9

<b>–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞</b>

–ó–Ω–∞—á–µ–Ω–∏—è ùëé –∏ ùëè (—Ä–∞–∑–¥–µ–ª—ë–Ω–Ω—ã–µ –ø—Ä–æ–±–µ–ª–æ–º).

–ï—Å–ª–∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø–∞—Ä –∑–Ω–∞—á–µ–Ω–∏–π ùëé –∏ ùëè, –¥–ª—è –∫–æ—Ç–æ—Ä—ã—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–∑–æ–≤–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ GCD –Ω–∞–∏–±–æ–ª—å—à–µ–µ, —Ç–æ –≤—ã–≤–µ–¥–∏—Ç–µ –ª—é–±—É—é –∏–∑ —Ç–∞–∫–∏—Ö –ø–∞—Ä.

```python
# –ù–∞–∏–±–æ–ª—å—à–µ–µ —á–∏—Å–ª–æ —à–∞–≥–æ–≤ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –ï–≤–∫–ª–∏–¥–∞
def inp_str(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line]
    return l

def print_list(l):
    for e in l:
        print(e,end=' ')

def GDCr(a,b,i):
    i+=1
    if a == 0 or b == 0:
        return max(a,b),i
    return GDCr(b,a%b,i)

# —Å–ª–∏—à–∫–æ–º –º–µ–¥–ª–µ–Ω–Ω—ã–π –¥–ª—è —Ç–µ—Å—Ç–∞ 3
def n_Euclid_slow(n):
    res = []
    for i in range(n+1):
        for j in range(i,n+1):
            res.append((i,j,GDCr(i,j,0)[1]))
    res.sort(key=lambda x: x[2])
    return ' '.join(str(x) for x in res[-1][:2])+f" n_steps={res[-1][2]}"

# –≤—Å–µ –µ—â–µ –º–µ–¥–ª–µ–Ω–Ω—ã–π
def n_Euclid_slow2(n):
    n_steps = 0
    pair = (1,n)
    for i in range(n+1):
        for j in range(n,i,-1):
            gdc_res = GDCr(i,j,0)
            if n_steps < gdc_res[1]:
                n_steps = gdc_res[1]
                pair = (i,j) 
    return pair, n_steps

# –º–∞–∫—Å–∏–º—É–º –≤—ã–∑–æ–≤–æ–≤ –¥–æ—Å—Ç–∏–≥–∞–µ—Ç—Å—è –¥–ª—è –¥–≤—Ö —á–∏—Å–µ–ª –§–∏–±–æ–Ω–∞—á—á–∏
def n_Euclid_fib(n):
    fib_1 = fib2(0)
    fib_2 = fib2(1)
    i = 2
    while fib_2 <= n:
        fib_new = fib2(i)
        if fib_new > n:
            return fib_1,fib_2
        fib_1 = fib_2
        fib_2 = fib_new
        i+=1
    return fib_1,fib_2


def fib2(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        fib_n2 = 0
        fib_n1 = 1
        fib_res = 1
        for _ in range(2,n):
            fib_n2 = fib_n1
            fib_n1 = fib_res
            fib_res = fib_n1+fib_n2        
    return fib_res


n=int(inp_str(0))
print_list(n_Euclid_fib(n))
```
