```python
def inp_str(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line]
    return l

def change(money):
    res = []
    c5 = money//5
    c1 = money - c5*5

    for i in range(c5+1):
        res.append( {1:c1+i*5,5:c5-i,10:0})
        rest_c5 = c5-i
        for j in range(2,rest_c5+1,2):
            res.append( {1:c1+i*5,5:c5-i-j,10:j//2})
    #res.sort(key=lambda x: x[1]+x[5]+x[10])
    return res

def str_change(dd):
    s = f"{dd[1]+dd[5]+dd[10]} "
    for c in dd:
        if dd[c]>0:
            s += ' '.join( str(c) for _ in range(dd[c]) ) + ' '
    while s[-1] == ' ':
        s = s[:-1]
    return s

money = int(inp_str(0))
result = change(money)
print(len(result))
for e in result:
    print(str_change(e))
```
