```python
def read_inp(file='input.txt'):
    with open(file,'r') as f:
        lines = f.readlines()
        n = int(lines[0])
        clicks = [ int(i) for i in lines[1].split(' ')  ]
        prices = [ int(i) for i in lines[2].split(' ')  ]
    return n, clicks, prices


n, clicks, prices = read_inp()

def MaxRevenue(n, clicks, prices):
    clicks.sort(reverse=True)
    prices.sort(reverse=True)
    revenue = 0
    for i in range(n):
        revenue += clicks[i]*prices[i]
    return revenue

print(MaxRevenue(n, clicks, prices))
```

ะะตัะฒะฐั ัััะพะบะฐ ัะพะดะตัะถะธั ััะธ ัะฐะทะดะตะปัะฝะฝัั ะฟัะพะฑะตะปะฐะผะธ ัะธัะปะฐ ๐, ๐ ะธ ๐ค (1โค๐โค10^3, 1โค๐โค10^5, 1โค๐คโค10^2).

ะะฐะปะตะต ะบะฐะถะดะฐั ะธะท ๐ ัััะพะบ ัะพะดะตัะถะธั ะฟะพ ะดะฒะฐ ัะฐะทะดะตะปัะฝะฝัั ะฟัะพะฑะตะปะพะผ ัะธัะปะฐ โ ๐_๐ ะธ ๐ค_๐ (1โค๐_๐โค10^2, 1โค๐ค_๐โค๐ค).

<b>ะคะพัะผะฐั ะฒัะฒะพะดะฐ</b>

ะัะฒะพะด ะดะพะปะถะตะฝ ัะพััะพััั ะธะท ะพะดะฝะพะณะพ ะทะฝะฐัะตะฝะธั ะฒะตะปะธัะธะฝั ๐๐๐ฅ_๐๐๐๐๐๐ก โ ะพัะฒะตั ะฝะฐ ะทะฐะดะฐัั.

```python
def read_inp(file='input.txt'):
    with open(file,'r') as f:
        lines = f.readlines()
        tmp = [int(i) for i in lines[0].split(' ')]
        adv = []
        for line in lines[1:]:
            adv.append(list(int(i) for i in line.split(' ')))
    return tmp[0], tmp[2], adv

def RevSlots(bbords, weeks, adv):
    adv.sort(key=lambda x: -x[0])
    slots = [0 for _ in range(weeks*bbords)]
    j=0
    for i in range(len(adv)):
        if j<len(slots):
            for _ in range(min(adv[i][1],len(slots)-j,weeks)):
                slots[j] = adv[i][0]
                j+=1
        else:
            return slots
    return slots

bbords, weeks, adv = read_inp()
print(sum(RevSlots(bbords, weeks, adv)))
```
