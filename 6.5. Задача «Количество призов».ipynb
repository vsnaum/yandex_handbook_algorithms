### <b>–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–æ–≤</b>

–ß—Ç–æ–±—ã –ø–æ—Ä–∞–¥–æ–≤–∞—Ç—å –∫–∞–∫ –º–æ–∂–Ω–æ –±–æ–ª—å—à–µ –¥–µ—Ç–µ–π, –≤–∞–º –ø–æ–Ω–∞–¥–æ–±–∏—Ç—Å—è –Ω–∞–π—Ç–∏ —Å–∞–º–æ–µ –±–æ–ª—å—à–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ ùëò k, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º —ç—Ç–æ –≤–æ–∑–º–æ–∂–Ω–æ.

<b>–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞</b>

–¶–µ–ª–æ–µ —á–∏—Å–ª–æ ùëõ.

<b>–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è: </b>

1‚â§ùëõ‚â§10^9

<b>–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞</b>

–ï–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–∞—è —Å—Ç—Ä–æ–∫–∞: –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ ùëò, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º ùëõ –º–æ–∂–Ω–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å –∫–∞–∫ —Å—É–º–º—É ùëò —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª.

```python
def inp_str(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line]
    return l

n = int(inp_str(0))

k=0
while n >= k*(k+1)/2:
    k+=1
print(k-1)
```
