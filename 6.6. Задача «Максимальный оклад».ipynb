–í —à–∞—Ö–º–∞—Ç–∞—Ö –∫–æ—Ä–æ–ª—å ‚Äî —Å–∞–º–∞—è –≤–∞–∂–Ω–∞—è —Ñ–∏–≥—É—Ä–∞. –ö–æ—Ä–æ–ª—å –º–æ–∂–µ—Ç —Å–¥–µ–ª–∞—Ç—å —Ö–æ–¥ –Ω–∞ –ª—é–±—É—é –∏–∑ 8 —Å–æ—Å–µ–¥–Ω–∏—Ö –∫–ª–µ—Ç–æ–∫ —à–∞—Ö–º–∞—Ç–Ω–æ–π –¥–æ—Å–∫–∏. –°–æ—Å–µ–¥–Ω–∏–º–∏ —Å—á–∏—Ç–∞—é—Ç—Å—è –∫–ª–µ—Ç–∫–∏ —Å–ª–µ–≤–∞, —Å–ø—Ä–∞–≤–∞, —Å–≤–µ—Ä—Ö—É, —Å–Ω–∏–∑—É –∏ 4 –ø–æ –¥–∏–∞–≥–æ–Ω–∞–ª–∏ –æ—Ç —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º–æ–π. –í —à–∞—Ö–º–∞—Ç–Ω–æ–π –ø–∞—Ä—Ç–∏–∏ –Ω–∞ –∏–≥—Ä–æ–≤–æ–º –ø–æ–ª–µ —Ä–æ–≤–Ω–æ –æ–¥–∏–Ω –∫–æ—Ä–æ–ª—å –∫–∞–∂–¥–æ–≥–æ —Ü–≤–µ—Ç–∞.

–†–∞—Å—Å–º–æ—Ç—Ä–∏ –∑–∞–¥–∞—á—É-–≥–æ–ª–æ–≤–æ–ª–æ–º–∫—É –æ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö —à–∞—Ö–º–∞—Ç–Ω—ã—Ö –∫–æ—Ä–æ–ª—è—Ö –Ω–∞ –¥–æ—Å–∫–µ –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞.

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º —à–∞—Ö–º–∞—Ç–Ω—É—é –¥–æ—Å–∫—É —Å ùëü —Å—Ç—Ä–æ–∫–∞–º–∏ –∏ ùëê —Å—Ç–æ–ª–±—Ü–∞–º–∏. –ù–∞–π–¥–∏—Ç–µ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ—Ä–æ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã—Ö –º–æ–∂–Ω–æ —Ä–∞–∑–º–µ—Å—Ç–∏—Ç—å –Ω–∞ –¥–æ—Å–∫–µ ùëü√óùëê, —Å –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —Å–ª–µ–¥—É—é—â–∏—Ö —É—Å–ª–æ–≤–∏–π:

–í –ª—é–±–æ–π –∫–ª–µ—Ç–∫–µ –Ω–∞ –¥–æ—Å–∫–µ –º–æ–∂–µ—Ç –Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –Ω–µ –±–æ–ª–µ–µ –æ–¥–Ω–æ–≥–æ –∫–æ—Ä–æ–ª—è.
–£ –ª—é–±–æ–≥–æ –∫–æ—Ä–æ–ª—è –µ—Å—Ç—å –ø–æ –∫—Ä–∞–π–Ω–µ–π –º–µ—Ä–µ –æ–¥–∏–Ω –≤–æ–∑–º–æ–∂–Ω—ã–π —Ö–æ–¥ –Ω–∞ —Å–≤–æ–±–æ–¥–Ω—É—é –∫–ª–µ—Ç–∫—É –Ω–∞ –¥–æ—Å–∫–µ.

<b>–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞</b>

–ï–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ ùëü –∏ ùëê.

<b>–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è: </b>

1‚â§ùëü,ùëê100.

<b>–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞</b>

–ù–∞–∏–±–æ–ª—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ—Ä–æ–ª–µ–π.

```python
def inp_list(line,fname='input.txt'):
    with open(fname,'r') as f:
        l = f.readlines()[line].split(' ')
        for i in range(len(l)):
            l[i] = int(l[i])
    return l


r,c = inp_list(0)


def kings(r,c):

    k = c*r
    minus_r = r//3
    minus_c = c//3

    if r%3!=0:
        minus_r +=1
    if c%3!=0:
        minus_c +=1

    k = k - minus_r*minus_c 
    return k

print(kings(r,c))
```
